```python
from google.colab import drive
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as plt
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/home3.jpg')
print(img.shape)
gImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#plt.imshow (gImg, cmap= 'gray')
cv2_imshow(gImg)
```

```python
img2 = img.copy()
```

```python
edge = cv2.Canny(gImg, 100, 200)
cv2_imshow(edge)
```

```python
lines = cv2.HoughLinesP(edge, 1, np.pi/180,200)
```

```python
lines = cv2.HoughLines(edge, 1, np.pi/180, 200)

for line in lines:
    rho, theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a*rho
    y0 = b*rho

    x1 = int(x0 + 1000*(-b))
    y1 = int(y0 + 1000*(a))
    x2 = int(x0 - 1000*(-b))
    y2 = int(y0 - 1000*(a))

    cv2.line(img2, (x1, y1), (x2, y2), (0,0,255), 2)

cv2_imshow(img2)
```

```python
lines = cv2.HoughLinesP(edge, 1, np.pi/180,200)

if lines is not None:
  for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(img2, (x1,y1), (x2,y2), (0,0,255), 2)

cv2_imshow(img2)
```
