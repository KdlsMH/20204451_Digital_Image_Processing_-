```python
from google.colab import drive
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as plt
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/lena256.jpg')
print(img.shape)
gimg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow (gimg, cmap= 'gray')
#cv2_imshow(gimg1)
```

```python
outimg = np.zeros(gimg.shape, dtype=np.uint8)
cv2_imshow(outimg)
```

```python
scaling = 3

for x in range(gimg.shape[0]):
    for y in range(gimg.shape[1]):
        h = int(y/scaling)
        w = int(x/scaling)
        if h < gimg.shape[0] and w < gimg.shape[1]:
            outimg[y,x] = gimg[h,w]
cv2_imshow(outimg)
```

```python
#outimg = cv2.resize(gimg, None, fx=3, fy=3, interpolation=cv2.INTER_NEAREST)
#outimg = cv2.resize(gimg, None, fx=3, fy=3, interpolation=cv2.INTER_LINEAR)
outimg = cv2.resize(gimg, None, fx=3, fy=3, interpolation=cv2.INTER_CUBIC)
cv2_imshow(outimg)
```

```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/lena256.jpg')
print(img.shape)
gimg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#plt.imshow (gimg, cmap= 'gray')
cv2_imshow(gimg)
```

```python
height, width = gimg.shape
print(height, width)
```

```python
M = cv2.getRotationMatrix2D((width/2, height/2), 45, 1)
print(M)
```

```python
outimg = cv2.warpAffine(gimg, M, (width, height))
cv2_imshow(outimg)
```
